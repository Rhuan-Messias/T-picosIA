```python
from ultralytics import YOLO
import torch
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")
```

```python
#Load Model
model = YOLO("yolo26n.pt")

# Train the model
results = model.train(
    data="../Dataset/cats.yaml",
    epochs=300,
    imgsz=640,

    # Performance
    batch=32,           # A4500 aguenta bem isso
    workers=8,

    # Otimização
    optimizer="AdamW",
    lr0=0.001,
    lrf=0.01,
    momentum=0.937,
    weight_decay=0.0005,

    # Warmup
    warmup_epochs=3,
    warmup_momentum=0.8,
    warmup_bias_lr=0.1,

    # Regularização / Augmentation
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,

    degrees=0.0,
    translate=0.1,
    scale=0.5,
    shear=0.0,

    fliplr=0.5,
    mosaic=1.0,
    mixup=0.1,

    # Treino
    patience=50,
    cos_lr=True,

    # Outros
    device=device,
    amp=True
)
```
